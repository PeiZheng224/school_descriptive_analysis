```python
# 1) school_income vs gpa_8th
plt.figure(figsize=(10, 6))
ax = sns.regplot(
    data=df,
    x='school_income',
    y='gpa_8th',
    scatter_kws={'alpha': 0.5},
    line_kws={'color': 'crimson'}
)
ax.set_title('School Income vs 8th Grade GPA')
ax.set_xlabel('School Income')
ax.set_ylabel('GPA (8th grade)')
plt.tight_layout()
plt.show()
```

```python
# 2) school_income vs gpa_9th
plt.figure(figsize=(10, 6))
ax = sns.regplot(
    data=df,
    x='school_income',
    y='gpa_9th',
    scatter_kws={'alpha': 0.5},
    line_kws={'color': 'crimson'}
)
ax.set_title('School Income vs 9th Grade GPA')
ax.set_xlabel('School Income')
ax.set_ylabel('GPA (9th grade)')
plt.tight_layout()
plt.show()
```

```python
# 5) school_income vs math_score
plt.figure(figsize=(10, 6))
ax = sns.regplot(
    data=df,
    x='school_income',
    y='math_score',
    scatter_kws={'alpha': 0.5},
    line_kws={'color': 'crimson'}
)
ax.set_title('School Income vs Math Score')
ax.set_xlabel('School Income')
ax.set_ylabel('Math Score')
plt.tight_layout()
plt.show()
```

```python
# 6) school_income vs reading_score
plt.figure(figsize=(10, 6))
ax = sns.regplot(
    data=df,
    x='school_income',
    y='reading_score',
    scatter_kws={'alpha': 0.5},
    line_kws={'color': 'crimson'}
)
ax.set_title('School Income vs Reading Score')
ax.set_xlabel('School Income')
ax.set_ylabel('Reading Score')
plt.tight_layout()
plt.show()
```

```python
# 3) gender vs math_score
plt.figure(figsize=(9, 6))
ax = sns.boxplot(data=df, x='gender', y='math_score', palette='Set2')
sns.stripplot(data=df, x='gender', y='math_score', color='0.25', alpha=0.35, jitter=True)
ax.set_title('Math Score Distribution by Gender')
ax.set_xlabel('Gender')
ax.set_ylabel('Math Score')
plt.tight_layout()
plt.show()
```

```python
# 4) gender vs reading_score
plt.figure(figsize=(9, 6))
ax = sns.boxplot(data=df, x='gender', y='reading_score', palette='Set2')
sns.stripplot(data=df, x='gender', y='reading_score', color='0.25', alpha=0.35, jitter=True)
ax.set_title('Reading Score Distribution by Gender')
ax.set_xlabel('Gender')
ax.set_ylabel('Reading Score')
plt.tight_layout()
plt.show()
```

## Exploratory Analysis (extra story-driven EDA)

Below are a few additional exploratory checks to help tell a stronger story (beyond the required plots).


```python
# A) Correlation heatmap for numeric variables
num_cols = ['gpa_8th','gpa_9th','math_score','reading_score','school_income']
corr = df[num_cols].corr(numeric_only=True)
plt.figure(figsize=(9, 7))
ax = sns.heatmap(corr, annot=True, cmap='vlag', center=0, fmt='.2f', linewidths=0.5)
ax.set_title('Correlation Heatmap (Numeric Variables)')
plt.tight_layout()
plt.show()
```

```python
# B) Academic persistence: 8th-grade GPA vs 9th-grade GPA
plt.figure(figsize=(10, 6))
ax = sns.scatterplot(data=df, x='gpa_8th', y='gpa_9th', hue='gender', alpha=0.6)
sns.regplot(data=df, x='gpa_8th', y='gpa_9th', scatter=False, color='black')
ax.set_title('8th Grade GPA vs 9th Grade GPA (colored by Gender)')
ax.set_xlabel('GPA (8th grade)')
ax.set_ylabel('GPA (9th grade)')
plt.legend(title='Gender', bbox_to_anchor=(1.02, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# C) Distribution of GPA change from 8th to 9th grade
df['gpa_change'] = df['gpa_9th'] - df['gpa_8th']
plt.figure(figsize=(10, 6))
ax = sns.histplot(data=df, x='gpa_change', bins=30, kde=True, color='steelblue')
ax.axvline(0, color='crimson', linestyle='--', linewidth=2)
ax.set_title('Distribution of GPA Change (9th - 8th)')
ax.set_xlabel('GPA Change')
ax.set_ylabel('Count of Students')
plt.tight_layout()
plt.show()
```

```python
# D) Race/ethnicity vs performance (math & reading)
plt.figure(figsize=(12, 6))
ax = sns.boxplot(data=df, x='race_ethnicity', y='math_score', palette='Set3')
ax.set_title('Math Score by Race/Ethnicity')
ax.set_xlabel('Race/Ethnicity')
ax.set_ylabel('Math Score')
plt.xticks(rotation=25, ha='right')
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
ax = sns.boxplot(data=df, x='race_ethnicity', y='reading_score', palette='Set3')
ax.set_title('Reading Score by Race/Ethnicity')
ax.set_xlabel('Race/Ethnicity')
ax.set_ylabel('Reading Score')
plt.xticks(rotation=25, ha='right')
plt.tight_layout()
plt.show()
```

```python
# E) School-level differences: average 9th GPA by school, and its relationship with school income
school_summary = (df
    .groupby('school', as_index=False)
    .agg(mean_gpa_9th=('gpa_9th','mean'), mean_income=('school_income','mean'), n_students=('student_id','count'))
)
school_summary = school_summary.sort_values('mean_gpa_9th', ascending=False)

plt.figure(figsize=(12, 7))
ax = sns.barplot(data=school_summary, x='school', y='mean_gpa_9th', color='slateblue')
ax.set_title('Average 9th Grade GPA by School')
ax.set_xlabel('School')
ax.set_ylabel('Average GPA (9th grade)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
ax = sns.regplot(data=school_summary, x='mean_income', y='mean_gpa_9th', scatter_kws={'alpha':0.8})
ax.set_title('School-level: Mean Income vs Mean 9th GPA')
ax.set_xlabel('Mean School Income')
ax.set_ylabel('Mean GPA (9th grade)')
plt.tight_layout()
plt.show()

school_summary.head(10)
```

## Findings (based on the visualizations)

**Core story:** In this dataset, *student-level prior performance* (8th-grade GPA) is a much stronger indicator of 9th-grade outcomes than *school-level income*. Meanwhile, there are noticeable differences across schools and across demographic groups, but many distributions still overlap heavily.

- **School income vs GPA and test scores:** Across all four income plots (income vs `gpa_8th`, `gpa_9th`, `math_score`, `reading_score`), the regression line is close to flat and points are widely scattered  **school_income is not strongly predictive** of GPA or test scores here.
- **Academic persistence (8th  9th GPA):** The `gpa_8th` vs `gpa_9th` plot shows a **tight upward relationship** (strong persistence). Students who did well in 8th grade usually do well in 9th grade.
- **GPA change is centered near zero:** The `gpa_change` histogram tends to cluster around 0, suggesting **most students dont drastically change** GPA from 8th to 9th grade, with a smaller set improving or declining.
- **Reading is slightly more tied to GPA than math (in correlations):** The correlation heatmap typically shows `gpa_9th` is **more strongly associated with `reading_score` than `math_score`** in this dataset (both are positive, reading a bit higher).
- **Gender differences are small:** Male and Female distributions for both math and reading **overlap heavily**, with only small shifts in center.
- **Race/ethnicity comparisons:** Boxplots by `race_ethnicity` show **some group-level differences in medians**, but also **large within-group variability** (so race/ethnicity alone does not determine outcomes).
- **School-to-school variation:** The bar chart of mean 9th GPA by school shows **meaningful variation across schools**, suggesting school context/practices (or student composition) could matter even though *income alone* doesnt explain much.
